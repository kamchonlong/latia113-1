```python
def a(x):
  result=0
  while result<10:
    result=result+x
    if result==5:
      return result
  return result
b=a(1)
c=a(2)
print(b)
print(c)
```

```python
def tf(value):
  return 3*value-2
test_values=[0,3,5,9,6,4]
scaled_values=[tf(value)for value in test_values]
print("Original values:",test_values)
print("scaled values:",scaled_values)
```
