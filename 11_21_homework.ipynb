```python
from bs4 import BeautifulSoup
import requests as rq
#we need to request a web site
url = 'https://github.com/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.146 Safari/537.36'}

response = rq.get(url, headers = headers, timeout=2)
html = response.content
#print(html)
soup = BeautifulSoup(html, "html.parser") # create a new bs4 object from the html data loaded
print(soup)
```

```python
soup = BeautifulSoup(html, "html.parser") # create a new bs4 object from the html data loaded
for script in soup(["script", "style"]): # remove all javascript and stylesheet code
  script.extract()
# get text
  text = soup.get_text()
# break into lines and remove leading and trailing space on each
  lines = (line.strip() for line in text.splitlines())
# break multi-headlines into a line each
  chunks = (phrase.strip() for line in lines for phrase in line.split(" "))
# drop blank lines
  texts = '\n'.join(chunk for chunk in chunks if chunk)
print(texts)
```

小型企業：新西蘭經濟的骨幹



從數據中我們可以看到，大多數新西蘭企業屬於小型企業。這些企業員工少，可能只有幾名員工，營業收入也相對較低，但它們在經濟中扮演著至關重要的角色。這些小型企業通常集中於本地市場，提供地區性服務或產品，並且靈活適應市場需求的變化。



例如，在服務業領域，許多家庭企業和創業型公司在這些數據中佔有一席之地，從餐飲、建築服務到美容和健康行業，都顯示出小型企業在創造就業機會和支持地方經濟發展方面的核心作用。



中型企業：穩定的經濟力量



進入中型企業範疇，我們看到這些企業在經濟中開始發揮更大作用。中型企業通常擁有較高的資金流動性和更穩定的財務狀況，能夠承接較大規模的項目並進行國際貿易。它們的經營範圍不僅限於本地市場，還可能進軍全球市場，尤其在製造業、農業和批發業中，中型企業成為了提升國家競爭力的關鍵。



例如，從調查數據中可以看出，某些行業如食品加工和精密製造，隨著企業規模的擴大，創造了更多的價值和就業機會，並且對經濟增長有著持久的正面影響。



大型企業：新西蘭的經濟引擎



大型企業在新西蘭的經濟中雖然數量較少，但它們擁有更強大的資本、更多的員工和較高的生產力。這些企業通常涉及跨國業務，參與全球貿易或在大規模基礎設施項目中發揮領頭羊作用。許多大型企業在技術創新、能源開發、交通物流等行業發揮著主導作用。



例如，從數據中我們可以看到，一些領先的大型企業，如製造業中的全球領航者或能源行業的巨頭，往往能夠穩定貢獻到國內生產總值（GDP）中，並且對國家財政和就業市場產生深遠影響。
