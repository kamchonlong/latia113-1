```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
# 創建數據集
data = {
'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81,
79, 84, 77, 80],
'Method': ['A', 'A', 'A', 'A', 'A', 'A', # 教學方法 A
'B', 'B', 'B', 'B', 'B', 'B', # 教學方法 B
'C', 'C', 'C', 'C', 'C', 'C'] # 教學方法 C

}
# 將數據轉換為 DataFrame
df = pd.DataFrame(data)
# 使用 statsmodels 中的 ols 方法進行單因子 ANOVA 分析
model = ols('Score ~ C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
# 輸出 ANOVA 表
print(anova_table)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import random
from scipy import stats

# Sample data (replace with your actual data)
numbers = [random.randint(0, 100) for _ in range(100)]
letters = [random.choice('ABC') for _ in range(100)]
data = {'numbers': numbers, 'letters': letters}
df = pd.DataFrame(data)

# Step 1: Calculate the frequency of each letter (for visualization)
letter_counts = df['numbers'].value_counts()

# Create the pie chart using the letter counts
plt.pie(letter_counts, labels=letter_counts.index, radius=1.5)
plt.title("Letter Distribution in the Data")
plt.show()

# Step 2: Perform ANOVA to test for differences in 'numbers' based on 'letters'
# Group the data by 'letters'
group_A = df[df['letters'] == 'A']['numbers']
group_B = df[df['letters'] == 'B']['numbers']
group_C = df[df['letters'] == 'C']['numbers']

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(group_A, group_B, group_C)

# Step 3: Interpret the results
print(f"ANOVA test statistic: {f_statistic}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("There is a significant difference between the groups.")
else:
    print("There is no significant difference between the groups.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import random
from scipy import stats

# Sample data (replace with your actual data)
numbers = [random.randint(0, 100) for _ in range(100)]
letters = [random.choice('ABC') for _ in range(100)]
data = {'numbers': numbers, 'letters': letters}
df = pd.DataFrame(data)

# Step 1: Calculate the frequency of each letter (for visualization)
letter_counts = df['letters'].value_counts()

# Create the pie chart using the letter counts
plt.pie(letter_counts, labels=letter_counts.index, radius=1.5)
plt.title("Letter Distribution in the Data")
plt.show()

# Step 2: Perform ANOVA to test for differences in 'numbers' based on 'letters'
# Group the data by 'letters'
group_A = df[df['letters'] == 'A']['numbers']
group_B = df[df['letters'] == 'B']['numbers']
group_C = df[df['letters'] == 'C']['numbers']

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(group_A, group_B, group_C)

# Step 3: Interpret the results
print(f"ANOVA test statistic: {f_statistic}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("There is a significant difference between the groups.")
else:
    print("There is no significant difference between the groups.")
```

圖表說明：



在這段代碼中，首先通過生成隨機的字母資料（‘A’, ‘B’, ‘C’）來創建一個 圓餅圖，顯示了每個字母出現的頻率。



	•	letter_counts = df['letters'].value_counts()：這行程式碼統計了字母 ‘A’, ‘B’, ‘C’ 出現的頻率，即每個字母在資料中出現的次數。

	•	plt.pie(letter_counts, labels=letter_counts.index, radius=1.5)：這行程式碼用來繪製圓餅圖，將每個字母對應的頻率顯示為圖中的區域。letter_counts.index表示每個字母，而 letter_counts 則是它們對應的數量。

	•	plt.title("Letter Distribution in the Data")：這行為圓餅圖設置了標題，描述圖表所表示的是字母在資料中的分佈情況。



這個圖表會幫助我們視覺化資料中字母的分佈，了解不同字母出現的頻率。這對於後續的ANOVA分析至關重要，因為它顯示了資料的分組（根據字母A、B、C分組）的均衡性或不均衡性。



ANOVA 結果解釋：



在這段程式碼中，我們進行了 單因素方差分析（ANOVA），目的是檢測基於字母（‘A’, ‘B’, ‘C’）分組的數字（numbers）是否存在顯著差異。



	•	ANOVA的假設：

	•	零假設 (H0)：字母A、B、C的數字均值沒有顯著差異。

	•	對立假設 (H1)：字母A、B、C的數字均值有顯著差異。

	•	stats.f_oneway(group_A, group_B, group_C)：這行程式碼執行了ANOVA檢驗，通過將每個字母組對應的數字（group_A, group_B, group_C）作為參數傳遞給 f_oneway() 函數，來檢驗這三個組別的數字是否存在顯著差異。

	•	f_statistic 和 p_value：

	•	F統計量（f_statistic）：衡量組間差異的程度。如果F值較大，表示組間差異較大。

	•	p值（p_value）：用來判斷結果是否顯著。如果 p 值小於預設的顯著性水平（通常是 0.05），則我們拒絕零假設，認為不同字母組之間的數字均值存在顯著差異。



結果的解釋：



	1.	如果 p 值 < 0.05：

	•	我們拒絕零假設，認為字母A、B、C組之間的數字均值存在顯著差異。這意味著這三個字母組在數字上的分佈並不相同，可能存在某些潛在的關聯或影響。

	2.	如果 p 值 >= 0.05：

	•	我們不能拒絕零假設，這意味著字母A、B、C組之間的數字均值沒有顯著差異。這表明，這三個組別的數字分佈在統計上是相似的，沒有顯著的差異。



圖表與結果之間的關係：



	•	圓餅圖顯示了不同字母的數量分佈情況，而ANOVA檢驗則分析了這些字母組在數字（numbers）上的均值是否存在顯著差異。圖表的分佈情況會對ANOVA結果產生影響：

	•	如果圖表顯示某些字母組的數量明顯多於其他字母組，這可能暗示數字的分佈存在某些偏差，從而可能影響ANOVA的結果。

	•	圓餅圖能夠提供一個直觀的視覺提示，告訴我們是否需要關注某些字母組的數量是否不均，這會影響後續的統計分析。



總結來說，圖表提供了數據分佈的視覺化，而 ANOVA檢驗則幫助我們在數字層面上評估這些分組是否有顯著差異。這兩者共同作用，有助於我們理解資料的結構，並為決策提供支持。
